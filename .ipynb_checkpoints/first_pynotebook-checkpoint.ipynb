```python
data = pd.read_csv('data.csv')
```

```python
data.head()
```

Left Back, with rating of over 60 rating, wage under < 50K

```python
data.columns
```

```python
data.Position.unique()
```

```python
leftbacks = data['Position'] == 'LB'
data[['Name', 'Club', 'Wage']][leftbacks]
```

```python
data['Wage'] = data['Wage'].apply(lambda x: int(re.findall('\d+', x)[0] + '000'))
```

```python
data[['Name', 'Club', 'Wage']]
```

```python
over60 = data['Wage'] <= 60000
overall = data['Overall'] >= 75

data[['Name', 'Club', 'Overall', 'Position', 'Wage']][over60 & leftbacks & overall]
```

```python
plt.hist(data['Overall'][leftbacks], bins=20)
plt.ylabel('Number of Players')
plt.xlabel('Left Back Ratings')
plt.show()
```

```python
rightbacks = data['Position'] == 'RB'
legend = ['Leftbacks', 'Rightbacks']

plt.hist([data['Overall'][leftbacks], data['Overall'][rightbacks]], color=['#c6e2ff', '#663366'])
plt.xlabel("Ratings")
plt.ylabel("Number of Players")
plt.legend(legend)
#plt.xticks(range(0, 7))
#plt.yticks(range(1, 20))
plt.title('Let\'s try comparing LB vs. RB Overall')
plt.show()
```

```python
df = data[['Nationality', 'Overall']]
df2 = data[['Nationality', 'Age']]
```

```python
country_avgs = df.groupby(['Nationality'], as_index=False).mean().round(decimals=1)
country_avgs2 = df2.groupby(['Nationality'], as_index=False).mean().round(decimals=1)
```

```python
ax = plt.gca()
country_avgs.sort_values('Overall', ascending=False).head(10).plot(kind='bar',x='Nationality',y='Overall',color='red', ax=ax)
country_avgs2.sort_values('Age', ascending=False).head(10).plot(kind='bar',x='Nationality',y='Age',color='blue', ax=ax)
plt.show()
```
